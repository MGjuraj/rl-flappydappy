```python
import os
import numpy as np
import cv2
import wrapped_flappy_bird as flappy_bird
from collections import deque
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, initializers
import matplotlib.pyplot as plt
```

```python
num_actions = 2

# initiate a game
game = flappy_bird.GameState()

# get the first state by doing nothing
do_nothing = np.zeros(num_actions)
do_nothing[0] = 1
print(do_nothing)
image, reward, terminal = game.frame_step(do_nothing)

print('shape of image:', image.shape)
print('reward: ', reward)
print('terminal: ', terminal)

plt.imshow(image.transpose([1, 0, 2]))
plt.show()
plt.close()

# Stored as example image for experimentation...
start_img = image
```

```python
for i in range(587):
    
    # choose a random action 
    action = np.random.choice(num_actions)
    
    # create the corresponding one-hot vector
    action_vec = np.zeros(num_actions)
    action_vec[action] = 1

    if action_vec[1] == 1:
        try:
            pre_jump_image
        except:
            print("first pre flap")
            pre_jump_image = image
    
    # take the action and observe the reward and the next state
    image, reward, terminal = game.frame_step(action_vec)
    
    if action_vec[1] == 1:
        try:
            post_jump_image
        except:
            print("first post flap")
            post_jump_image = image
    
    if i < 5:
        print(action)
        print(action_vec)
        print('reward: ', reward)
        print('terminal: ', terminal)
```
